# Psychrometric Chart Example

```python
import matplotlib.pyplot as plt

from pythermalcomfort.models import pmv_ppd_iso
from pythermalcomfort.plots import ranges_psychrometric

ax, _ = ranges_psychrometric(
    pmv_ppd_iso,
    fixed_params={
        "tr": 25.0,
        "met": 1.0,
        "clo": 0.61,
        "vr": 0.1,
        "wme": 0.0,
    },
    thresholds=[-0.5, 0.5],
    t_range=(10, 36),
    w_range=(0.0, 0.03),
    w_step=5e-4,
    # plot_kwargs={"band_alpha": 0.35, "line_color": "k", "cmap": "viridis"},
)
plt.show()
```
